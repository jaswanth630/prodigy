```python
year_column = '2022'
selected_countries=df.head(6)
plt.figure(figsize=(15, 8))
plt.bar(selected_countries['Country Name'],selected_countries[year_column])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Distribution of Population in {year_column}')
plt.xticks(rotation=45)
plt.show()
```
